```python
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
```

```python
train_data=pd.read_csv('../data/kaggle_house_pred_train.csv')
figure=plt.figure()
sns.pairplot(x_vars=['OverallQual','GrLivArea','YearBuilt','TotalBsmtSF'], y_vars=['SalePrice'], data=train_data, dropna=True)
```

```python
train_data = pd.read_csv('../data/kaggle_house_pred_train.csv')
test_data = pd.read_csv('../data/kaggle_house_pred_test.csv')
all_features = pd.concat((train_data.iloc[:, 1:-1], test_data.iloc[:, 1:]))

count = all_features.isnull().sum().sort_values(ascending=False)
ratio = count / len(all_features)

pd.concat([count, ratio], axis=1, keys=['count', 'ratio'])
```
